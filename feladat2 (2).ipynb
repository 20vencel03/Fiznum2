# Oszcillátorok numerikus vizsgálata

## Először ismételjük át az alapokat:

Oszcillátoroknak olyan dinamikai rendszereket nevezünk, melyek mozgása az idő periódikus vagy kvázi-periódikus függvényeként írható le. Az egydimenziós oszcillátor mozgásegyenlete

$$ m \frac{\dd^2x}{\dd t^2} = - \frac{\dd V(x)}{\dd x} $$

alakban írható. Egy szabad oszcillátor teljes energiája mindaddig megmarad, amíg a rendszerben nem hat súrlódási erő (csillapított oszcillátor), illetve nem hatunk rá külső erővel (gerjesztett oszcillátor). Az oszcillátor teljes energiája a kinetikus és a potenciális energia összege, azaz

$$ E_{\textrm{tot}} = \frac{m v^2}{2} + V(x)$$

### Harmonikus oszcillátor

Amennyiben a $V(x)$ potenciál négyzetes, abban az esetben harmonikus oszcillátorról beszélünk:

$$\begin{eqnarray}
    V(x) & = & \frac{1}{2}kx^2 \\
    m \frac{\dd^2x}{\dd t^2} & = & - k x.
\end{eqnarray}$$

A harmonikus oszcillátor mozgásegyenlete analitikusan is megoldható. A megoldás

$$ x(t) = A \cos( \omega t + \varphi ) $$

alakú, ahol $A$ az amplitúdó - vagyis a legnagyobb kitérés -, $\omega$ a körfrekvencia és $\varphi$ a kezdőfázis $t = 0$-ban. A harmonikus oszcillátor körfrekvenciáját a potenciálban szereplő $k$ állandó és az $m$ tömeg jellegű mennyiség határozza meg:

$$ \omega = \sqrt{\frac{k}{m}} = \frac{2\pi}{T}, $$

ahol $T$ a pediódusidő, melyből $f = 1/T$ módon kaphatjuk meg a frekvenciát.

## Anharmonikus oszcillátor

Az anharmonikus abban különböznek a harmonikus oszcillátortól, hogy a mozgásukat meghatározó potenciál nem négyzetes alakú. Néhány lehetséges anharmonikus potenciál:

* Negyedrendű potenciál

$$ V(x) = \frac{1}{4}k_4 x^4 $$

* Lennard-Jones potenciál, ld. [a wikipédián](https://en.wikipedia.org/wiki/Lennard-Jones_potential)

$$ V(x) = V_0 \left[ \left( \frac{x_0}{x} \right)^{12} - 2 \left( \frac{x_0}{x} \right)^{6} \right] $$

## Csillapított és gerjesztett oszcillátor

A csillapított és/vagy gerjesztett oszcillátorok mozgásegyenletében megjelenik egy sebességtől függő és egy további, többnyire időfüggő külső erőtag is:

$$ m \frac{\dd^2x}{\dd t^2} = - \frac{\dd V(x)}{\dd x} - c \frac{\dd x}{\dd t} + F(t) $$

Csillapított és szinuszosan gerjesztett harmonikus oszcillátorok viselkedése analitikusan is egyszerűen vizsgálható. Az alábbi feladatokban ezért ezektől eltérő rendszereket vizsgálunk numerikus módszerek segítségével.

Miután importáltuk a scipy.integrate csomagból az RK45-t, ilymódon tudjuk használni: 

rk45 = RK45( 'fun', 't0', 'y0', 't_bound', 'max_step=inf', 'rtol=0.001', 'atol=1e-06' )

Itt a 'fun' függvényt fentebb, az a) részben  definiáltuk, 't0' a kezdő időpont, 'y0' a kezdő állapot.

Lényeges eltérés az odeint függvénytől, hogy az RK45 függvényhíváskor még csak egy objektumot kapunk eredményül. Ennek .step metódusát tudjuk az integrálási lépésekre használni, a .t, .y attribútumaiként pedig megkapjuk a változók értékeit. Tehát ha az RK45 eredményét a fentiek szerint eltároljuk, akkor a léptetés rk45.step() utasítással végezhető, az eredményeket pedig rk45.t, rk45.y értékeiként kapjuk meg.

A mozgást leíró differenciálegyenletek harmonikus potenciál esetén:

$$\dot x = v$$
$$m \dot v = -kx$$

ahol $x$ a kitérés az egyensúlyi helyzettől az idő függvényében, az $\dot x$ pedig az idő szerinti deriváltat jelöli.

A mozgást leíró differenciálegyenletek Lennard-Jones potenciál esetén:

$$\dot x = v$$
$$m \dot v = -12 V_0 \cdot \frac{x_0^{12}}{x^{13}}+12V_0 \cdot \frac{x_0^6}{x^7}$$

a fenti kifejezésben $V_0$ és $x_0$ a potenciált jellemző paraméterek.

```python
# a megoldás helye

%pylab inline
from scipy.integrate import RK45
```

A megoldandó differenciálegyenletek az a) részben a következők:



```python
def harmonikusmozgas(t, ikszve):
    
    k=1
    
    xpont = ikszve[1]         #Itt adom meg, hogy x derivaltja a sebesség
    vpont = -k*ikszve[0]              #Itt adom meg a ható erő nagyságát
    
    
    return [xpont, vpont]
```

```python
harmonikusmozgas(2,[2,3])
```

```python
t0=0
x0=1

kezdetix = 1
ikszve0=[kezdetix,0]  # kezdő időpont és állapot: 3 kitérés, és 0 sebesség
```

```python
#A példából kimásolt és átírt kód

atol=1e-6; rtol=1e-7                          # Abszolút és relatív pontosság mgadása. ezek praktikusan jó értékek.
rk=RK45(harmonikusmozgas,t0,ikszve0,inf,atol=atol,rtol=rtol)  # A diffegyenlet rendszer megoldásának előkészítése:
           # Itt mondjuk meg a diffegyenletrendszert tartalmazó függvény nevét, és a kezdő időt és állapotot.
           # E függvényhívás hasonlít ahhoz, ahogy az odeint függvényt használtuk,
           # csakhogy az odeint egyből elvégezte a megoldást egy neki megadott időintervallumon,
           # ez az RK45 viszont csak egy objektumot ad vissza, amit eltárolunk, pl. *rk* néven.
           # Azt aztán az alábbi módon tudjuk léptetésre, és az állapot kiolvasására használni.
           # Másrészt 4. argumentumként egy időkorlátot lehet neki megadni, aminek a szerepe később lesz világos.
           # Még egy eltérés, hogy a *harmonikusmozgas* függvényben a *t* és az állapot változó sorendje ellentétes kell legyen
           # az odeint használata esetében szokásoshoz képest.
tlist=[]   # Ezekbe fogjuk gyűjteni az adatokat
ikszvelist=[]
ikszve=ikszve0     # a kezdő időpontot és állapotot betöltjük, hogy az első while lefutás is tudja ellenőrizni
#t1=380

while len(tlist)<=150:
#while ikszve[0]>1:         # addig léptessen, amíg az állapot zv vektorának z komponense pozitív
                       # így amikor leáll, az utolsó z érték 0 vagy negatív lesz már (és az lesz az első ilyen)
    t=rk.t             # Így olvassuk ki a pillanatnyi időt (kezdetben ez az RK45-nek megadott t0-t adja vissza)
    tlist.append(t)    # eltároljuk a listába
    ikszve=rk.y            # Az y objektumváltozóból lehet kiolvasni a pillanatnyi állapotvektort.
    ikszvelist.append(ikszve)  # Ezt is eltároljuk.
    rk.step()          # A step metódus pedig egy Runge-Kutta lépést végez.
                       # Ha túllépett az RK45 4. argumentumában megadott időkorláton, akkor nem léptet többet.
                       # Ha ilyen korlátozást nem akarunk, akkor ott inf-et adunk be.
# Látszik, hogy míg az odeint automatikusan végigszalad a neki megadott időintervallumon,
# a .step metódus egyszerre csak egy lépést végez. A lépés nagyságát ez is automatikusan választja, hibabecslés alapján.
```

```python
ikszvearray=array(ikszvelist)          # A plot-beli indexeléshez kétindexes array-vé alakította.
plot(tlist,ikszvearray[:,0],"o-")  # Ábrázolta.
axhline(0,color="r")           # hogy látsszon a z=0 egyenes (lehet a grid() utasítással is)
```

## <font color='green5'>5. Szorgalmi feladatrész</font>

Programozzunk be kaotikus oszcillátort!$^1$ Egy szinuszosan gerjesztett oszcillátor esetében a kaotikus viselkedés feltétele, hogy a visszatérítő erő ne lineáris függvénye legyen a kitérésnek. Ábrázoljuk a kitérés és a sebesség időfüggését, valamint a teljesítményspektrumot! Futassuk az integrálást legalább 1 millió lépésig, és ábrázoljuk a rendszer Poincaré-metszetét, azaz a sebességet a kitérés függvényében az $\omega t = n \cdot 2 \pi, n = 1, 2, 3, ... $ feltétel mellett!

```python
# a megoldás helye

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the Lorenz system equations
def lorenz(state, t, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Set the system parameters
sigma = 10.0
rho = 28.0
beta = 8.0/3.0

# Set the initial conditions
state0 = [1.0, 1.0, 1.0]

# Set the time points for the simulation
t = np.linspace(0, 50, 5000)

# Solve the differential equations
states = odeint(lorenz, state0, t, args=(sigma, rho, beta))

# Plot the results
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(states[:,0], states[:,1], states[:,2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the system parameters
sigma = 10.0
rho = 28.0
beta = 8.0/3.0

# Set the fixed state variables
y = 0.0
z = 0.0

# Set the initial condition for the x variable
x0 = 1.0

# Set the number of time steps to simulate
n_steps = 10000

# Initialize an array to store the oscillator values
x = np.zeros(n_steps+1)
x[0] = x0

# Simulate the one-dimensional Lorenz system
for i in range(n_steps):
    dxdt = sigma * (y - x[i])
    x[i+1] = x[i] + dxdt

# Plot the results
plt.plot(x)
plt.xlabel('Time')
plt.ylabel('Oscillator value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the control parameter
r = 3.6

# Set the initial condition
x0 = 0.1

# Set the number of time steps to simulate
n_steps = 10000

# Initialize an array to store the oscillator values
x = np.zeros(n_steps+1)
x[0] = x0

# Simulate the logistic map
for i in range(n_steps):
    x[i+1] =  r* x[i] * (1 - x[i])

# Plot the results
plt.plot(x[:100])
plt.xlabel('Time')
plt.ylabel('Oscillator value')
plt.show()
```
